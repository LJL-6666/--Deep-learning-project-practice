```python
from PIL import Image     # 图片处理
import numpy as np        # 数据处理
from matplotlib import pyplot as plt  # 画板

Im0=Image.open('./lake.jpg')   # 读取图片
Im0=np.array(Im0)/255          # 将图⽚转为数组，并将像素值处理到0-1之间
p,q,d=Im0.shape               # 矩阵的长度
plt.imshow(Im0)               # 图片展示
```

```python
Im1=np.zeros([p,q,d])  # 建⼀个和样本⻓度⼀样的3维数组
for x in range(p):
    for y in range(q):
        x0=np.min([x+50,p-1])    # x轴缩减50
        y0=np.min([y+100,q-1])     # y轴缩减100
        Im1[x0,y0,:]=Im0[x,y,:]    # 设置坐标
plt.imshow(Im1)      # 展示
```

```python
Im2=np.zeros([p,q,d])    # 建⼀个和样本⻓度⼀样的3维数组
for x in range(p):
    Im2[p-x-1,:,:]=Im0[x,:,:]    # 图片翻转
plt.imshow(Im2)                  # 展示
```

```python
Im3=np.zeros([p,q,d])     # 建⼀个和样本⻓度⼀样的3维数组
for x in range(p):
    for y in range(q):
        x0=int(x*0.8+y*0.2);x0=np.min([p-1,x0])   # 旋转，缩放
        y0=int(x*0.2+y*0.8);y0=np.min([q-1,y0])   # 旋转，缩放
        Im3[x0,y0,:]=Im0[x,y,:]                   # 设置坐标
plt.imshow(Im3)
```

```python
Im4=np.zeros([p,q,d])           # 建⼀个和样本⻓度⼀样的3维数组
for i in range(d): Im4[:,:,i]=Im0[:,:,d-i-1]     # 像素点取样
plt.imshow(Im4)             # 展示
```

```python
Im5=np.zeros([2*p,q,d])    # 建⼀个3维数组
Im6=np.zeros([2*p,q,d])    # 建⼀个3维数组

Im5[range(p),:,:]=Im0     # 以下底为对称轴
Im5[range(p,2*p),:,:]=Im2    # 镜面翻转

Im6[range(p),:,:]=Im2      # 以上底为对称轴
Im6[range(p,2*p),:,:]=Im0    # 镜面翻转

fig,ax=plt.subplots(1,2)   # 建立画布
ax[0].imshow(Im5)          # 展示
ax[1].imshow(Im6)
```

```python
Im7=np.zeros([q,p,d])    # 建立一个3维数组
Im8=np.zeros([q,p,d])    # 建立一个3维数组
for i in range(p): 
    Im7[:,i]=Im0[i,:]      # 左转
    Im8[:,i]=Im0[p-i-1,:]   # 右转

fig,ax=plt.subplots(1,2)  # 建立画布
ax[0].imshow(Im7)         # 展示
ax[1].imshow(Im8)
```
